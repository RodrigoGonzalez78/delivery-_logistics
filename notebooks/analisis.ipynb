```python
data = df["delivery_partner"].value_counts()
colors = sns.color_palette('pastel')[0:len(data)]

plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distyribnucion de Partners', fontsize=16)
plt.show()
```

```python
data = df["package_type"].value_counts()
colors = sns.color_palette('pastel')[0:len(data)]

plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribucin de tipos de paquetes', fontsize=16)
plt.show()
```

```python
data = df["vehicle_type"].value_counts()
colors = sns.color_palette('pastel')[0:len(data)]

plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribucion de tipos de vehiculos', fontsize=16)
plt.show()
```

```python
partners = df['delivery_partner'].unique()

for partner in partners:
    plt.figure(figsize=(12, 6))
    subset = df[df['delivery_partner'] == partner]
    
    sns.countplot(
        data=subset,
        x='package_type',
        hue='vehicle_type',
        palette='pastel'
    )
    
    plt.title(f'{partner.title()} - Package & Vehicle Distribution', fontsize=14)
    plt.xlabel('Package Type')
    plt.ylabel('Count')
    plt.legend(title='Vehicle Type', loc="upper right")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
numeric_cols = ['delivery_cost', 'delivery_rating', 'distance_km', 'package_weight_kg']

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.boxplot(
        data=df,
        x='delivery_partner',
        y=col,
        ax=axes[i],
        palette='pastel'
    )
    axes[i].set_title(f'{col} Distribution by Delivery Partner', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel(col)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
numeric_cols_df = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols_df].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))

# Subplot 1: Distancia
plt.subplot(1, 2, 1)
sns.boxplot(x=df['distance_km'], color='skyblue')
plt.title('Distribucin de Distancia (km)')

# Subplot 2: Peso del paquete
plt.subplot(1, 2, 2)
sns.boxplot(x=df['package_weight_kg'], color='salmon')
plt.title('Distribucin de Peso del Paquete (kg)')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Histograma de Distancia
sns.histplot(df['distance_km'], kde=True, bins=30, color='blue')
plt.title('Histograma de Distancias')
plt.xlabel('Kilmetros')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(df["delivery_time_hours"], bins=20, alpha=0.5, label="Delivery", color="blue")
plt.hist(df["expected_time_hours"], bins=20, alpha=0.5, label="Expected", color="orange")
plt.xlabel("Time (hours)")
plt.ylabel("Frequency")
plt.title("Delivery vs Expected Time Distribution")
plt.legend()
plt.show()
```
